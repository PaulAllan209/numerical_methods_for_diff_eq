```python
import numpy as np
import pandas as pd

from decimal import *
getcontext().prec = 100
```

```python
xi = None
yi = None

global h1
h1 = 0.25

decimal_places = '0.000001'

x_range = 5
```

```python
def given_func(x, y):
    x = Decimal(str(x)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    y = Decimal(str(y)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    
    main_func = Decimal('-2')*(x**3) + Decimal('12')*(x**2) - Decimal('20')*x + Decimal('8.5')
    
    return Decimal(str(main_func)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)


def method_func(xi, yi, h=h1):
    xi = Decimal(str(xi)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    yi = Decimal(str(yi)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    h = Decimal(str(h)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    
    return Decimal(str(yi + (given_func(xi, yi))*h)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
```

```python
# appending x values to array
list_x = []
i = 0
while i <= x_range:
    list_x.append(i)
    i+=h1


data_arr = np.zeros((len(list_x), 1), dtype=np.float64)
data_arr[:,0] = np.transpose(list_x)
```

```python
# making y values
list_y = [1]
```

```python
for i in range(len(list_x) - 1):
    
    yi = method_func(list_x[i], list_y[i])
    
    list_y.append(yi)


data_arr = np.append(data_arr, np.transpose([list_y]), axis=1)
```

```python
df = pd.DataFrame(data_arr, columns=['x', 'y'])
df
```
