```python
import numpy as np
import pandas as pd

from decimal import *
getcontext().prec = 100
```

```python
xi = None
yi = None
k1 = None
k2 = None

global h1
h1 = 0.25

decimal_places = '0.000001'

x_range = 5
```

```python
def given_func(x, y):
    x = Decimal(str(x)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    y = Decimal(str(y)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    
    main_func = Decimal('-2')*(x**3) + Decimal('12')*(x**2) - Decimal('20')*x + Decimal('8.5')
    
    return Decimal(str(main_func)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)


def method_func(xi, yi, k1, k2, h=h1):
    xi = Decimal(str(xi)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    yi = Decimal(str(yi)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    k1 = Decimal(str(k1)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    k2 = Decimal(str(k2)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    h = Decimal(str(h)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    
    return Decimal(str(yi + (((Decimal('0.5')*k1)) + ((Decimal('0.5')*k2)))*h)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)

def return_k1(xi, yi):
    return given_func(xi, yi)

def return_k2(xi, yi, k1, h=h1): 
    xi = Decimal(str(xi)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    yi = Decimal(str(yi)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    k1 = Decimal(str(k1)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    h = Decimal(str(h)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    
    x = xi + h
    y = yi + (k1*h)
    
    return given_func(x, y)
```

```python
# appending x values to array
list_x = []
i = 0
while i <= x_range:
    list_x.append(i)
    i+=h1


data_arr = np.zeros((len(list_x), 1), dtype=np.float64)
data_arr[:,0] = np.transpose(list_x)
```

```python
# making y values
list_y = [1]
list_k1 = [0]
list_k2 = [0]
```

```python
for i in range(len(list_x) - 1):
    k1 = return_k1(list_x[i], list_y[i])
    k2 = return_k2(list_x[i], list_y[i], k1)
    
    yi = method_func(list_x[i], list_y[i], k1, k2)
    
    list_y.append(yi)
    list_k1.append(k1)
    list_k2.append(k2)


data_arr = np.append(data_arr, np.transpose([list_y]), axis=1)
data_arr = np.append(data_arr, np.transpose([list_k1]), axis=1)
data_arr = np.append(data_arr, np.transpose([list_k2]), axis=1)
```
