```python
import numpy as np
import pandas as pd

from decimal import *
getcontext().prec = 100
```

```python
xi = None
yi = None

k1 = None
k2 = None
k3 = None
k4 = None

global h1
h1 = 0.25

decimal_places = '0.000001'

x_range = 5
```

```python
def given_func(x, y):
    x = Decimal(str(x)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    y = Decimal(str(y)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)

    return Decimal(str(y*(x**Decimal('2')) - Decimal('1.1')*y)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)

def method_func(yi, k1, k2, k3, k4, h=h1):
    yi = Decimal(str(yi)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    k1 = Decimal(str(k1)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    k2 = Decimal(str(k2)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    k3 = Decimal(str(k3)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    k4 = Decimal(str(k4)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    h = Decimal(str(h)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    
    
    return Decimal(str(yi + Decimal(1/6)*(k1 + Decimal(2)*(k2) + Decimal(2)*(k3) + k4)*h)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)

def return_k1(xi, yi):
    return given_func(xi, yi)

def return_k2(xi, yi, k1, h=h1):
    xi = Decimal(str(xi)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    yi = Decimal(str(yi)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    k1 = Decimal(str(k1)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    h = Decimal(str(h)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    
    x = xi + Decimal(1/2)*(h)
    y = yi + (Decimal(1/2)*(k1)*(h))
    
    return given_func(x, y)

def return_k3(xi, yi, k1, k2, h=h1):
    xi = Decimal(str(xi)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    yi = Decimal(str(yi)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    k1 = Decimal(str(k1)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    k2 = Decimal(str(k2)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    h = Decimal(str(h)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    
    x = xi + Decimal(1/2)*(h)
    y = yi + Decimal(1/2)*(k2)*(h)
    
    return given_func(x,y)
    
def return_k4(xi, yi, k1, k2, k3, h=h1):
    xi = Decimal(str(xi)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    yi = Decimal(str(yi)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    k1 = Decimal(str(k1)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    k2 = Decimal(str(k2)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    k3 = Decimal(str(k3)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    h = Decimal(str(h)).quantize(Decimal(decimal_places), rounding=ROUND_HALF_UP)
    
    x = xi + h
    if Decimal('23202.077082437500000000') == x:
        x = Decimal('23202.07708')
    
    y = yi + (k3)*(h)
    
    
    return given_func(x,y)
```

```python
# appending x values to array
list_x = []
i = 0
while i <= x_range:
    list_x.append(i)
    i+=h1

  
data_arr = np.zeros((len(list_x), 1), dtype=np.float64)
data_arr[:,0] = np.transpose(list_x)
```

```python
# making y values
list_y = [1]
list_k1 = [0]
list_k2 = [0]
list_k3 = [0]
list_k4 = [0]
```

```python
for i in range(len(list_x) - 1):
    k1 = return_k1(list_x[i], list_y[i])
    k2 = return_k2(list_x[i], list_y[i], k1)
    k3 = return_k3(list_x[i], list_y[i], k1, k2)
    k4 = return_k4(list_x[i], list_y[i], k1, k2, k3)
    
    
    yi = method_func(list_y[i], k1, k2, k3, k4)
    
    list_y.append(yi)
    list_k1.append(k1)
    list_k2.append(k2)
    list_k3.append(k3)
    list_k4.append(k4)
    

data_arr = np.append(data_arr, np.transpose([list_y]), axis=1)
data_arr = np.append(data_arr, np.transpose([list_k1]), axis=1)
data_arr = np.append(data_arr, np.transpose([list_k2]), axis=1)
data_arr = np.append(data_arr, np.transpose([list_k3]), axis=1)
data_arr = np.append(data_arr, np.transpose([list_k4]), axis=1)
```

```python
df = pd.DataFrame(data_arr, columns=['x', 'y', 'k1', 'k2', 'k3', 'k4'])
df
```
